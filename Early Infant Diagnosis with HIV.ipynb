# Early Infant Diagnosis with HIV Analysis

```python
import pandas as pd

# df is your dictionary of DataFrames with sheet names as keys
df = pd.read_excel('HIV_Early_Infant_Diagnosis_2023.xlsx', sheet_name='Data', header=0)

# Set the header to the first row
df.columns = df.iloc[0]

# Drop the first row as it is now the header
df = df[1:]

# Display the modified DataFrame
print(df)
```

```python
i = 'Reported number of infants born to pregnant women living with HIV who received a virological test for HIV within 2 months of birth'
data = df[df['Indicator'] == i]

data = data[data['UNICEF Region'] == 'West and Central Africa']
data
```

```python
data.tail()
```

```python
# drop column of lower upper that have missing values
data=data.drop(['Lower', 'Upper'], axis=1)
data=data.reset_index(drop=True)
data
```

```python

# df is your DataFrame and 'Value' is the column you want to convert
data['Value'] = data['Value'].str.replace(',', '').astype(int, errors='raise')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')

# contains country names
countries = data['Country/Region'].unique()

# Create subplots
fig, axs = plt.subplots(len(countries) // 3 + 1, 3, figsize=(15, 25))
fig.suptitle('Early Infant Diagnosis with HIV in West and Central Africa Countries')


# Plot each country
for i, country in enumerate(countries):
    country_data = data[data['Country/Region'] == country]
    row = i // 3
    col = i % 3
    axs[row, col].plot(country_data['Year'], country_data['Value'])
    axs[row, col].set_title(country)
    axs[row, col].set_xlabel('Year')

# Adjust layout for better visualization
plt.tight_layout()
plt.show()
```

```python
# Group by 'Year' and 'Country/Region', calculate the mean of 'Value'
wc = data.groupby(['Year', 'Country/Region'])['Value'].mean()

# Reset the index to make 'Year' and 'Country/Region' regular columns
wc = wc.reset_index()
```

```python

# create a bar plot of the mean temperature change over time
plt.figure(figsize=(15, 6))
c=wc.groupby(['Year'])['Value'].mean()
df = c.reset_index()
plt.bar(df['Year'], df['Value'])
plt.title('Early Infant Diagnosis with HIV by Year in West and Central Africal', fontsize=16)
plt.xlabel('Year')
plt.xticks(rotation=90)
plt.ylabel('Average Value Diagnosis ')
plt.grid(True)
plt.show()
```

```python

# create a line plot of the mean temperature change over time
plt.figure(figsize=(15, 6))
c=wc.groupby(['Year'])['Value'].mean()
c.plot()
plt.title('Early Infant Diagnosis by Year in West and Central Africa', fontsize=16)
plt.xlabel('Year')
plt.xticks(rotation=90)
plt.ylabel('Average Value Diagnosis ')
plt.grid(True)
plt.show()
```

## Apply Machine Learning to predict the next year for West and Central Africa Infant Diagnosis

```python
df
```

```python
import numpy as np

X=df['Year']
X=np.array(X).reshape(-1,1)
X.shape
```

```python
y=df['Value']
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=44)

# Fit the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions on both training and test sets
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Plotting training set
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.scatter(X_train, y_train, label='Actual (Training)')
plt.plot(X_train, y_pred_train, color='b', label='Predicted (Training)')
plt.title('Infant Diagnosis West and Central Africa (Training Set)')
plt.xlabel('Year')
plt.ylabel('Value Diagnosis')
plt.legend()

# Plotting test set
plt.subplot(1, 2, 2)
plt.scatter(X_test, y_test, label='Actual (Test)')
plt.plot(X_test, y_pred_test, color='r', label='Predicted (Test)')
plt.title('Infant Diagnosis West and Central Africa (Test Set)')
plt.xlabel('Year')
plt.ylabel('Value Diagnosis')
plt.legend()

plt.tight_layout()
plt.show()

# Plotting residuals
plt.figure(figsize=(15, 5))

plt.scatter(y_pred_test, y_pred_test - y_test, color='r')
plt.hlines(y=0, xmin=min(y_test), xmax=max(y_test), color='k', linewidth=2)
plt.title('Residuals Plot')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')

plt.show()
```

```python
print('Coefficient of determination(R-Squared):', model.score(X_train, y_train))
print('Intercept:', model.intercept_)
print('slope:', model.coef_)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Predictions on the test set
y_pred = model.predict(X_test)

# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)

# Calculate mean absolute error
mae = mean_absolute_error(y_test, y_pred)

# Print the results
print("Mean Squared Error:", round(mse, 2))
print("Mean Absolute Error:", round(mae, 2))
```
